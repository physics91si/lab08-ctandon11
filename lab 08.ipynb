```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
#load data
data=np.loadtxt('droptower_vdata.txt', unpack=True)
```

```python
#position
position=integrate.cumtrapz(data[1],data[0],initial=0)+30
```

```python
#acceleration
acceleration=np.gradient(data[1], data[0])
```

```python
#graph
f, axarr = plt.subplots(3)
axarr[0].plot(data[0],position,color='r',)
axarr[0].set_title('position')
axarr[1].plot(data[0],data[1],color='b')
axarr[1].set_title('velocity')
axarr[2].plot(data[0],acceleration)
axarr[2].set_title('acceleration')
plt.subplots_adjust(top=0.85)
plt.show()
```
